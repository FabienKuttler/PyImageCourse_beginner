# 10. Morphology solutions

````python
import numpy as np
import matplotlib.pyplot as plt
import skimage
````

1. Import the morphology module of skimage

````python
import skimage.morphology
````

2. Import the image from ```../Data/myoblast.tif```

````python
image = skimage.io.imread('../Data/myoblast.tif')
````

3. Create a new variable ```image_green``` containing the green channel image and plot it

````python
image_green = image[:,:,1]
plt.imshow(image_green, cmap = 'gray')
plt.show()
````

4. Use the Otsu thresholding method to create a mask containing the structures visible in the image and plot it

````python
th_li = skimage.filters.threshold_otsu(image_green)
mask = image_green > th_li
plt.imshow(mask, cmap = 'gray')
plt.show()
````

5. Erode and then dilate the image with a disk of radius 5. Plot the result

````python
disk = skimage.morphology.disk(5)

image_erode = skimage.morphology.binary_erosion(mask, selem=disk)
image_dilate = skimage.morphology.binary_dilation(image_erode, selem=disk)

plt.imshow(image_dilate, cmap = 'gray')
plt.show()
````

6. Now do the opposite: first erosion then dilation

````python
disk = skimage.morphology.disk(5)

image_dilate = skimage.morphology.binary_dilation(mask, selem=disk)
image_erode = skimage.morphology.binary_erosion(image_dilate, selem=disk)

plt.imshow(image_erode, cmap = 'gray')
plt.show()
````

6. In the [skimage.morphology module](http://scikit-image.org/docs/dev/api/skimage.morphology.html) , try to find a function that is creating a skeleton of the mask of point #6, i.e. that is thinning to the maximum all the structures. Plot the result

````python
image_skeleton = skimage.morphology.skeletonize(image_erode)
````

````python
plt.subplots(figsize=(5,5))
plt.imshow(skimage.morphology.skeletonize(image_skeleton))
plt.show()
````
