# 5. Loops and conditions

We have seen in the previous chapter that we can define structures that can contain a lot of information. What should we do now if we want to do some operation for example on each element of a list? Of course we could manually treat each element but obviously that's not a good solution. What we can use are so called loops that automate the access to all those elements.

## 5.1 Simple looping over a list

Let's create a list:

````python
mylist = [8,3,9,20,27]
````

Let's imagine now that we want to calculate the square of each of these numbers. In English, one would say that "**for** each element **in** mylist, calculate the square of the element".

Luckily the Python syntax is designed to be intuitive and the Python code that does this is:

````python
for x in mylist:
    x**2
````

x just stands for the currently selected element from mylist. Note that:
1. The for loop ends with :
2. The content of the loop is **INDENTED**

You can also note that when we execute the cell nothing happens. This is because no graphical output is generated fro for loops. If we want to see the actual value we have to use the print() function:

````python
for x in mylist:
    value_squared = x**2
print(value_squared)
````

Only the last value is printed because we put the print() function outside the loop. If we want to see each value we have to **indent** the print() call so that it is included in the loop:

````python
for x in mylist:
    value_squared = x**2
    print(value_squared)
````

## 5.3 Looping using a range

Often we don't want to loop over the content of a list but just want to do some operation N times or for indexes from 0 to N. To do that, we can use the range() built-in function that just does this: it provides numbers within a certain range. The function doesn't really produce a list per se but can be used as if it were one. For example:

````python
for x in range(8):
    print(x)
````

Note that as always the first index is not 0 but 8. Of course we could use these indexes to access specific parts of a list. Coming back to the previous example, we might want to calculate the square only of the three first numbers:

````python
for x in range(3):
    value_squared = mylist[x]**2
    print(value_squared)
````

With ```mylist[x]``` we simply use the numbers generated by range() as indexes of our list.

Another type of loop available in most programming languages is the while loop. We are not going to used it today, but you should definitely learn about it.

## 5.3 Conditional statements

When automating operations typically with a for loop we sometimes want to make sure that we only do the operation in some particular case. We say "if this condition is true, then do the operation". The way to write this in Python is with an if statement. The syntax is similar to the for loop with the if statement followed by : and the operation indented.

````python
a = 3
````

````python
if a < 5:
    print('yes')
````

If we want to execute another operation if the condition is not met we can use an elif statement:

````python
a = 1
if a < 2:
    print('smaller than 2')
elif a < 5:
    print('smaller than 5')
````

Note that the if statement stops the **first** time a condition is verified. Finally we can use the else statement if none of the previous conditions is realized.

````python
a = 10
if a < 2:
    print('smaller than 2')
elif a < 5:
    print('smaller than 5')
else:
    print('larger than 5')
````

## 5.4 Mixing for and if

Let's come back to our initial for loop, and say now that we only want to calculate the square of numbers larger than 10. We can integrate our if condition in the for loop:

````python
mylist = [8,3,9,20,27]

for x in mylist:
    
    if x > 10:
    
        value_squared = x**2
        print(value_squared)
````

Note how we create different levels of indentation. Everything which belongs to the for loop is indentedn once, including the if statement. But everything that belongs to the if statement is also indented relative to it, and so ends up **indented twice**.
